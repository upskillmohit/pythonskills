```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sumeven(listnum):
  sum = 0
  for i in listnum:
    if i%2 ==0:
      sum = sum + i

  return sum

print(sumeven([22,54,99,45,33,21,4,2,8,6,7]))


```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def sqrtnum(listnum):
  sqrtnum = []
  for i in listnum:
    sqrtnum.append(i**2)

  return sqrtnum

print(sqrtnum([1,2,3,4,5]))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def checkPrime(num):
  if num == 1:
    return False
  if num == 2:
    return True
  div = 2
  while div < num/2:
    if num % div == 0:
      return False
    div = div + 1

  return True

if checkPrime(17):
  print("Prime")
else:
  print("Not Prime")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

class fibsequence:
  def __init__(self,range):
    self.range = range
    self.count = 0
    self.first = 0
    self.second = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.range:
      raise StopIteration
    else:
      previous = self.first
      self.first = self.second
      self.second = self.first + self.second
      self.count = self.count + 1
      return previous

fibrange = fibsequence(5)

for num in fibrange:
  print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


def powers_of_num(num, powers):
  count = 0
  while count < powers :
    yield num**count
    count += 1

for num in powers_of_num(2,5):
  print(num)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string

def readbyline(filename):

  openfile = open(filename, "r")
  # Read the first line
  yield openfile.readline()
  line = openfile.readline()
  while line:
    yield line
    line = openfile.readline()  # Read the next line
  # Close the file
  openfile.close()

for line in readbyline("MessagingPatterScenarios.txt"):
   print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
def sortlist(tuplelist):
  tuplelist.sort(key = lambda x : x[1])
  return tuplelist

sortedmylist = sortlist([(1, 2, 3), (4, 1, 6), (9, 10, 12), (13, -3, 7)])
print(sortedmylist)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def removevowels(stringwithvowels):
  vowels = ['a','e','i','o','u','A','E','I','O','U']
  stringwithoutvowels = ""
  for chr in stringwithvowels:
    if chr not in vowels:
      stringwithoutvowels = stringwithoutvowels + chr
  return stringwithoutvowels

print(''.join(filter(removevowels,"Print string without vowels")))
```
