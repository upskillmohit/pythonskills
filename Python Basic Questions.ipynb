  2. What is an interpreter in Python?
    - It is a program that reads and executes codes direcly without converting it in to machine language. Examples of Python interpreters are CPython, PyPy, IronPython etc.

  3. What are pre-defined keywords in Python?
     - Predifined keywords hold a special meaning and purpose e.g. break, continue, false, true etc. They are used to write programming logic.

  4. Can keywords be used as variable names ?
   - No we cannot use keywords as variable names. Python interpreter will throw exception if they are used to name variables.

  5. What is mutability in Python?
     - It is the ability of an object to be modified after creation. Examples for mutable objects are Lists, Dictionaries, Set.

  6. Why are lists mutable, but tuples are immutable?
     - Lists are mutable because elements can be added, removed and changed after the list is created. It is useful when there is a need to modify exising data frequently as the object share the same memory.
     - Tuples are immutable because elements of a tuple cannot be modified after the tuple is created. It ensures safety and performance.

  7. What is the difference between “==” and “is” operators in Python?

    - “==” operator is used to compare values of two objects. e.g x==y
    - "is" operator is used to check identity of objects. It checks whether values at both the sides of the operator point to same object. e.g. a is b

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types.

def conversion(price):
  print("Type of Price before converting to int ", type(price))
  price = int(price)
  print("Type of price after converting to int", type(price))
  print("Converted to int", price)
  print("Type of price before converting to float ", type(price))
  price = price * 5.5
  print("Type of price after converting to float ", type(price))
  print("Converted to Float", price)
  print("Type of price before converting to bool ", type(price))
  price = bool(price)
  print("Type of price after converting to bool ", type(price))
  print("Converted to Bool", price)

try:
  price = input("Enter price: ")
  conversion(price)
except:
  print("Invalid Input")
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.
sum = 0
for num in range(1,51):
   if (num % 2) == 0:
      sum = sum + num

print(f"Sum of all even numbers between 1 and 50 is {sum}")
```

```python
# 14. Write a program to reverse a string using a while loop.
def reverse_string(stringtorev):
  last = len(stringtorev)
  print("Reversed String: ", end="")
  while last > 0:
      last = last - 1
      print(stringtorev[last], end="")


try:
  stringtorev = input("Enter a string to reverse: ")
  reverse_string(stringtorev)
except:
  print("Invalid Input")

```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
def findfact(num):
  fact = 1
  while num > 0:
    fact = fact * num
    num = num - 1
  print (f"Factorial of number : {fact}")


number = int(input("Enter a number: "))
findfact(number)

```
